```python
input_columns = ['age', 'bp', 'sg', 'al', 'su','pcv', 'wc']
```

```python

X = dataset[input_columns]
```

```python
y = dataset['classification']
```

```python
print("Features (X):")
print(X.head())
```

```python
print("\nTarget (y):")
print(y.head())
```

```python
from imblearn.over_sampling import SMOTE
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

imputer = KNNImputer(n_neighbors=5)
X_imputed = imputer.fit_transform(X_scaled)

X_final = pd.DataFrame(scaler.inverse_transform(X_imputed), columns=X.columns)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42, k_neighbors=1)
X_resampled, y_resampled = smote.fit_resample(X_final, y)

print("Before SMOTE:", y.value_counts())
print("After SMOTE:", y_resampled.value_counts())
```

```python
# Remove whitespace and tab characters
y = y.str.strip()

print("Unique classes after cleaning:", y.unique())
print(y.value_counts())
```

```python
smote = SMOTE(random_state=42, k_neighbors=1)
X_resampled, y_resampled = smote.fit_resample(X_final, y)

print("After SMOTE:", y_resampled.value_counts())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Logistic Regression
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_resampled, y_resampled)
y_pred = log_reg.predict(X_test)

print("Logistic Regression Results:")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

# Random Forest
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_resampled, y_resampled)
y_pred_rf = rf.predict(X_test)

print("\nRandom Forest Results:")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))

# Plot confusion matrix for RF
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest Confusion Matrix")
plt.show()
```

```python
# Train RF without SMOTE
rf_no_smote = RandomForestClassifier(n_estimators=200, random_state=42)
rf_no_smote.fit(X_train, y_train)
y_pred_no_smote = rf_no_smote.predict(X_test)

print("\nRandom Forest without SMOTE:")
print(classification_report(y_test, y_pred_no_smote))
```

```python
import pickle

# Save the trained Random Forest model
with open("rf_model.pkl", "wb") as f:
    pickle.dump(rf, f)

print("✅ Model saved as rf_model.pkl")

# ------------------------------
# Later, load the model back
with open("rf_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)

# Test the loaded model
y_pred_loaded = loaded_model.predict(X_test)
print("\nLoaded model accuracy:", accuracy_score(y_test, y_pred_loaded))
```
