```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/chennai_rainfall_2019_2023_mm.csv")
```

```python
dataset
```

```python
input = dataset[["Year",	"Month",	"Temperature (°C)",	"Humidity (%)",	"Wind Speed (km/h)"]]
```

```python
output = dataset[["Rainfall (cm)" , "Rainfall (mm)"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(input,output,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin = LinearRegression()
```

```python
lin.fit(xtrain,ytrain)
```

```python
pr = lin.predict(xtest)
```
