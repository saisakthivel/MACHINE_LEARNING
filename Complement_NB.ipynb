```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('/content/loan_approval_dataset.csv')
```

```python
dd=pd.get_dummies(df,columns=['education','self_employed','loan_status'], dtype=int, drop_first=True)
```

```python
final_data=dd.rename(columns={'loan_status_Rejected':'loan_status',
                              'education_Not Graduate':'Graduate',
                              'self_employed_Yes':'Employment'})
```

```python
x=final_data[["no_of_dependents",
    "income_annum",
    "loan_amount",
    "loan_term",
    "cibil_score",
              "residential_assets_value",
    "commercial_assets_value",
    "luxury_assets_value",
    "bank_asset_value",
    "Graduate",
    "Employment"]]
y=final_data["loan_status"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

# Replace negative values in 'residential_assets_value' with 0
x_train['residential_assets_value'] = x_train['residential_assets_value'].clip(lower=0)
x_test['residential_assets_value'] = x_test['residential_assets_value'].clip(lower=0)
```

```python
from sklearn.naive_bayes import ComplementNB
```

```python
nb=ComplementNB()
```

```python
traine=nb.fit(x_train,y_train)
```

```python
prediction = traine.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy= accuracy_score(prediction,y_test)
```
