```python
import pandas as pd
```

```python
ds = pd.read_csv("/content/loan_approval_dataset.csv")
```

```python
ds
```

```python
dfe = pd.get_dummies(ds,columns =["education" , "self_employed" , "loan_status"] , dtype = int , drop_first=True)
```

```python
dfe
```

```python
newds = dfe.rename(columns = {"education_Not Graduate" : "education" , 	"self_employed_Yes" : "employment"	, "loan_status_Rejected" : "loan_status"})
```

```python
newds
```

```python
input = newds[["no_of_dependents" ,	"income_annum" ,	"loan_amount"	, "cibil_score"	, "residential_assets_value" ,	"commercial_assets_value"	, "luxury_assets_value"	, "bank_asset_value" ,	"education"	, "employment"]]
```

```python
output = newds[["loan_status"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(input , output , test_size =0.20 , random_state=42)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Rejected', 'Approved'], yticklabels=['Rejected', 'Approved'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

print("Classification Report:\n")
print(classification_report(y_test, y_pred))
```
