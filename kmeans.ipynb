```python
import pandas as pd
```

```python

dataset = pd.read_csv('/content/India-2010.csv')
```

```python
dataset
```

```python
# for k-means we dont take independent and dependent
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Select features (M and F populations)
X = dataset[["M", "F"]]
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
dataset["Cluster"] = kmeans.fit_predict(X)
```

```python
# Show clustered data
print(dataset)
```

```python
# Plot clusters
plt.figure(figsize=(8,6))
plt.scatter(dataset["M"], dataset["F"], c=dataset["Cluster"], cmap="viridis", s=100)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            color="red", marker="X", s=200, label="Centroids")
plt.xlabel("Male Population")
plt.ylabel("Female Population")
plt.title("K-Means Clustering of Age Groups")
plt.legend()
plt.show()
```
