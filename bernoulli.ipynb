```python
import pandas as pd
```

```python
dataset = pd.read_csv('/content/loan_approval_dataset.csv')
```

```python
dataset
```

```python
df = pd.get_dummies(dataset,columns=['education','self_employed','loan_status'],dtype=int, drop_first=True)
```

```python
df
```

```python
independent = df[['loan_amount']]
dependent = df[['loan_status_Rejected']]
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
nb = BernoulliNB()
nb.fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
```
